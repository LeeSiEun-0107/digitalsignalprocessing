```python
#[ ,     ]
import matplotlib.pyplot as plt

df_15=pd.read_csv('C:/Users/eos82/Desktop/SW/world-happiness/2015.csv')
df_16=pd.read_csv('C:/Users/eos82/Desktop/SW/world-happiness/2016.csv')
df_17=pd.read_csv('C:/Users/eos82/Desktop/SW/world-happiness/2017.csv')
df_18=pd.read_csv('C:/Users/eos82/Desktop/SW/world-happiness/2018.csv')
df_19=pd.read_csv('C:/Users/eos82/Desktop/SW/world-happiness/2019.csv')

# 15    
print(df_15.groupby('Country')['Happiness Score'].mean())
# 15    
print(df_15.groupby('Region')['Happiness Score'].mean())
# 15      
print(df_15.groupby('Region')['Country'].nunique())

# 16     ,     
print(df_16.groupby('Country')['Happiness Score'].mean())
#print(df_16.groupby('Region')['Happiness Score'].mean())
print(df_16.groupby('Region')['Country'].nunique())

# 17      
print(df_17.groupby('Country')['Happiness Score'].mean())

# 18     
print(df_18.groupby('Country')['Happiness Score'].mean())

# 19     
print(df_19.groupby('Country')['Happiness Score'].mean())

# 16       
print(df_16.groupby('Country')['Happiness Score'].mean().idxmin())
print(df_16.groupby('Country')['Happiness Score'].mean().idxmax())

# 17       
print(df_17.groupby('Country')['Happiness Score'].mean().idxmin())
print(df_17.groupby('Country')['Happiness Score'].mean().idxmax())

# 18       
print(df_18.groupby('Country')['Happiness Score'].mean().idxmin())
print(df_18.groupby('Country')['Happiness Score'].mean().idxmax())

# 19       
print(df_19.groupby('Country')['Happiness Score'].mean().idxmin())
print(df_19.groupby('Country')['Happiness Score'].mean().idxmax())
```

```python
#       
import matplotlib.pyplot as plt
df_15_mean=df_15.mean()
df_16_mean=df_16.mean()
df_17_mean=df_17.mean()
df_18_mean=df_18.mean()
df_19_mean=df_19.mean()

df_mean=pd.concat([df_15_mean,df_16_mean,df_17_mean,df_18_mean,df_19_mean],axis=1,join='inner')
df_mean.columns = ['2015','2016','2017','2018','2019']
print(df_mean.plot(kind='bar', title='Happiness Score and factors contributing to happiness (2015 ~ 2019)', figsize=(12, 4), legend=True, fontsize=12))
```

```python
#    
import matplotlib.pyplot as plt
plt.suptitle("Change of Happiness Score(2016~2019)")
df_mean_change_hp=df_mean.loc['Happiness Score']
df_mean_change_hp.plot()
```

```python
#  ()  (scatter )
import matplotlib.pyplot as plt
scatter_plot=plt.figure()
plt.suptitle("Scatter plot of Health vs Happiness Score(2016~2019)")
plt.subplots_adjust(hspace=0.8)
axes2=scatter_plot.add_subplot(2,2,1)
axes2.scatter(df_16['Health (Life Expectancy)'],df_16['Happiness Score'],s=3)
axes2.set_title("2016")
axes2.set_xlabel('Health (Life Explectancy)')
axes2.set_ylabel('Happiness Score')

axes2=scatter_plot.add_subplot(2,2,2)
axes2.scatter(df_17['Health (Life Expectancy)'],df_17['Happiness Score'],s=3)
axes2.set_title("2017")
axes2.set_xlabel('Health (Life Explectancy)')
axes2.set_ylabel('Happiness Score')

axes2=scatter_plot.add_subplot(2,2,3)
axes2.scatter(df_18['Health (Life Expectancy)'],df_18['Happiness Score'],s=3)
axes2.set_title("2018")
axes2.set_xlabel('Health (Life Explectancy)')
axes2.set_ylabel('Happiness Score')

axes2=scatter_plot.add_subplot(2,2,4)
axes2.scatter(df_19['Health (Life Expectancy)'],df_19['Happiness Score'],s=3)
axes2.set_title("2019")
axes2.set_xlabel('Health (Life Explectancy)')
axes2.set_ylabel('Happiness Score')
```
